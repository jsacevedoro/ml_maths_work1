## Linear Model with Least Squares approach

In the first place, we propose the following linear model: $$ \hat{Y} = \theta_1 X_1 + \theta_2 X_2 + \theta_3 X_3 + \theta_4 X_4 $$

Where each feature $X$ means:

$ X_1 $ : Feature associated to the intercept, by default it is a m-dimensional vector consist of $\textbf{ones}$   

$ X_2 $ : Oil Price  

$ X_3 $ : Silver Price  

$ X_4 $ : Gold Price 

In the next steps, we build the matrix $X = [X_1 \hspace{0.4cm} X_2 \hspace{0.4cm} X_3 \hspace{0.4cm} X_4]$

Note that $X\in \mathbb{R}^{m\times n}$, where $m=365$, the numbers of samples and $n=4$, the number of feature plus the intercept. 

Next, we are going to build $X$ matrix, according to the data set. Also, it is defined $Y$ as an n-dimensional column vector which conatins the real bitcoin prices over the year.

```julia
# Useful constants
m,n = size(df);

# X definition
X_1 = ones(m);
X = [ X_1 Array(df[:,2:4])];

# Y Definition
Y = Array( df[:,"Bitcoin_price"] )

# Verify its dimenstions
println("X dimension: ",size(X))
println("Y dimension: ",size(Y))
```

Note that with the matrix notation we can write the predicted values by the model as

$$ \hat{Y} =  X\theta$$

where $\theta = (\theta_1 \hspace{0.4cm}\theta_2 \hspace{0.4cm} \theta_3 \hspace{0.4cm} \theta_4)^T$

Now, we are looking forward solving the Least Squares problem, so the objective function turns to be the Cost Function, given by, 

\begin{align}
C(\theta) &= \frac{1}{n} \sum_{i=1}^m (\hat{y}_i - y_i)^2 \\
          &= \frac{1}{n}(\hat{Y}-Y)^T(\hat{Y}-Y) \\
          &= \frac{1}{n}(X\theta - Y)^T(X\theta - Y) \\
          &= \frac{1}{n}||X\theta - Y||^2
\end{align}

Thus, the optimization problem is

$$argmin  \left\{ C(\theta) \hspace{0.3cm} \big| \hspace{0.3cm} \theta \in \mathbb{R}^4 \right\} = 
argmin  \left\{ \frac{1}{n}(X\theta - Y)^T(X\theta - Y) \hspace{0.3cm} \big| \hspace{0.3cm} \theta \in \mathbb{R}^4 \right\}
= argmin \left\{ \frac{1}{n}||X\theta - Y||^2 \hspace{0.3cm} \big| \hspace{0.3cm} \theta \in \mathbb{R}^4 \right\} $$ 

It's not difficult to see that this is a convex problem, then it has a unique solution, we will use the Normal Equation approach. Thus the solution $\theta^*$ is given by
$$\theta^* = X^+Y $$
Where $X^+$ is the Moore-Penrose pseudo-inverse of $X$.

Each component of $\theta^*$ is called the $\textbf{slope}$ or coeficient associated to it's feature. 

```julia
# Calculating the Slopes using the Normal Equation
θ = pinv(X)*(Y); 
print("Intercept \t \t $(θ[1]) \nOil price slope:\t $(θ[2]) \nSilver price slope: \t $(θ[3]) \nGold price slope: \t $(θ[4])")
```

With the explicit values of $\theta^*$ that we have just found, we can calculate the values predicted or fitted by the model.

```julia
# Calculating fitted or predicted values by the model 
Ŷ = X*θ;
print("Dimension of Ŷ: $(size(Ŷ))")
```

The following graphic shows the actual prices compared to the model-adjusted prices

```julia
plot(df[:,1],[Y Ŷ],  linewidth=2,
    label=["Actual Price" "Fitted by model"], legend_position=(0,1), 
    ylabel="Bitcoin Price", xlabel="Date" )
```

## Regression Analysis

### Slopes
From the value of the intercept one can infer that if suddenly the price of commodities decline near to 0, the bitcoin price would increase to 161700 dollars. So, in a world where real materials are not worth, perhaps the virtual cash would increase it's worth.

From the gold and oil slopes, we can see a similar behavior as that observed in the correlation analysis. The bitcoin and gold prices have an inverse relation, while the oil price is somehow proportional to the bitcoin's one.

### R Squared
The coefficient of determination $R^2$ is given by 

$$R^2 = 1 - \dfrac{\sum_{i=1}^m \left( y_i - \hat{y}_i\right)^2}
                {\sum_{i=1}^m \left( y_i - \overline{y} \right)^2} = 1 - \frac{SS_{residuals}}{SS_{total}}$$

Where $SS_{residuals}$ is the sum of squares over residuals and $SS_{total}$ the Total Sum of Squares.

We can write the $R^2$ in vectorial way as 
        $$R = 1 - \frac{\lvert \lvert Y-\hat{Y} \lvert \lvert ^2 }{ \lvert \lvert Y-\overline{Y} \lvert \lvert ^2}$$
        
where $\overline{Y}$ is an m-dimensional vector consisting of the mean value of $Y$ in its components.

The $R^2$ is a measure of the linear relationship between the independant variables as a whole and the dependant one. If it is close to 1 it means that the variance of Bitcoin price is explained by the commodity's prices. In multiple linear regression, the $R^2$ tends to increase when considering more variables, this would not be desirable, so it is common to compute the adjusted R squared, which penalizes the ammount of variables. It's formula is given by
$$R^2_{adj} = 1 - \frac{m-1}{m-n-1}(1-R^2)$$

```julia
# Computing the R squared
Ȳ = mean(Y)*ones(m);
R² = 1 - ((Y-Ŷ)'*(Y-Ŷ)) / ((Y-Ȳ)'*(Y-Ȳ));

# R cuadrado ajustrado 
Rₐ = 1 - ((m-1)/(m-n-1)) * (1-R²)

println("R² \t \t", R²)
println("Adjusted R² \t", Rₐ)
```

This means that about $27\%$ of the variance over the bitcoin's price can be explained by the prices of commodities. The remmaining percentage of the bitcoin's price behavior do not depend on the price of oil,silver and gold. This might be due to the residuals are not normal distribuited, as it is shown bellow, using a QQPlot.

```julia
using StatsPlots
qqplot(Y,Ŷ, "Residuals QQPlot")
```

This could mean that this is not a robust model, in the next analysis of p-values and t-values we will watch it clearly.

### t-values and p-values

```julia
lm(@formula(Bitcoin_price ~ Oil_price+Silver_price+Gold_price) , df)
```

The coefficient t-value is a measure of how many standard deviations our coefficient estimate is far away from 0. Note that the t-value associated to the silver price is really close to 0, which means that there is no relation between silver price and bitcoin price. According to the p-value of the silver price we can state, with 87 percent confidence, that there is no relation between that commodity and the bitcoin's price. On the other hand, one can state with more than $5\%$ of significance that the price of the oil and gold has an impact on the bitcoin price.
