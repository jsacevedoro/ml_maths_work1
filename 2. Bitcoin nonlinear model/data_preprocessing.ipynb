```python
import pandas as pd 

# Loading the data
df = pd.read_csv("BTC-USD.csv")
display(df)
```

```python
# Selecting columns of interest
df = df.loc[:,['Date','Close']]
# Rename the principal column
df.rename(columns={'Close':'Bitcoin_price'}, inplace=True)
# Fill missing values with interpolation
df.Bitcoin_price.interpolate(inplace=True)
# Set date as index 
df.loc[:,'Date'] = df.loc[:,'Date'].astype('datetime64[ns]')
df.set_index(['Date'], inplace=True)

display(df)
```

```python
# Cutting the data according to the date 

import datetime

first_date = datetime.datetime(2022, 3, 14)
df = df[first_date:]
df.plot()
```
