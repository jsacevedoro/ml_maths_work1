# Bitcoin price prediction using Least Squares for non-linear models

```julia
# Required packages

using DataFrames # To manipulate dataframes
using CSV # To Handle csv files
using Plots # To make plots
using Statistics # Satatistics calculations
using LinearAlgebra # To do some calculations
```

## Introduction
Given the historical data of bitcoin daily price in US-Dollars, we want to predict the next week daily price. To achieve this goal, we propose two non-linear models based on Least Squares, training with diferent time-lapses and testing the models over the last known week.

## Data Exploration
The data were taken from https://finance.yahoo.com/quote/BTC-USD/history/, on March 29, 2021. To see the details of data preprocessing go to the file in this same directory called  <data_proprocessing.ipynb>

We have the time series of the daily bitcoin's price. Let's take a look on the first data frame named "df_march".

```julia
# Loading data
df_march = CSV.read("df_march.csv",DataFrame);
# Displaying the first rows
display(df_march[1:5,:])
# Ploting the time series
plot(df_march[:,"Date"],df_march[:,"Bitcoin_price"], xlabel="Date", ylabel="USD", label="Bitcoin Price")
```

We also have a different dataframe with the bitcoin price over the two last weeks

```julia
# Loading data
df_2 = CSV.read("df_two_weeks.csv",DataFrame);
# Displaying the first rows
display(df_2[:,:])
# Ploting the time series
plot(df_2[:,"Date"],df_2[:,"Bitcoin_price"], xlabel="Date", ylabel="USD", label="Bitcoin Price")
```

Note that the two graphics show different patterns, this will be crucial when choosing functions for non-linear models.

## Models

### Parabola
Just watching the shape of the curve of the last two weeks we propose a non-linear model which is a second order polynomial , as shown as follows
    $$ \hat{y} = \alpha_1 + \alpha_2 x^2 $$ 
where $\hat{y}$ represents the bitcoin price which is dependent of $x$, the time in days.

We are going to adjust the model with the data of the first week and test them with the values of the second week. The exact dates are.

| Purpose | Date Range | 
| --- | --- |
| Adjust model | 14-03 to 21-03 |
| Testing | 22-03 to 28-03 |

So, with the eight first values of the "df_2" dataframe we will find the optimum parameters $\alpha_1,\alpha_2,\alpha_3$
that minimize the cost function 
$$C(\alpha) = \frac{1}{n}\sum_{i=1}^n (\hat{y}_i-y_i)^2 = \frac{1}{n}\sum_{i=1}^n (\alpha_1 + \alpha_2 x^2_i-y_i)^2 $$

```julia
# Define the x vector which represent 8 days 
x = 1:8
# Build the matrix according to the model
A = [ ones(8) x.^2]
```

```julia
# get the prices of the bitcoin for the days we are interested in
y = df_2[1:8,2]
```

We will use the Normal Equation to get the $\alpha$ parameters. For more detailed explanation watch the point 3. 

```julia
α = pinv(A)*y
```

Now we define the function $f$ associated to the model using the $\alpha$ parameters. 

```julia
f(x) = α[1] + α[2]*x.^2

z = Array(1:15)

plot(z, f.(z), color="green", label="Model", linewidth=2)
scatter!(x,df_2[1:8,2], color="red", label="Adjusted Nodes")
plot!(z,df_2[:,2], color="darkblue", label="Real values", legend_position=(0.3,0.9))
```

The root mean squared error is computed as
$$RMSE = \sqrt{MSE} = \sqrt {\frac{1}{n}\sum_{i=1}^n (\hat{y}_i-y_i)^2} = 
\sqrt{\frac{1}{n}\sum_{i=1}^n \lvert\lvert \hat{y}_i-y_i \lvert \lvert ^2} $$

Because the square root is a monotonic increasing function, it turns to be that minimizing the RMSE and MSE are equivalent optimization problems, the latter (MSE) is the objective function solving the Least Squares. The RMSE benefit is that it is given in the same scale of the variable $y$, in our case, dollars.

In the next code cell we are computing the RMSE comparing the results of the model and the real values over the last week known.

```julia
# Compute the root mean squared error 

# Get the actual prices if bitcoin during the last week
Y = df_2[9:end,2]
# Time lapse in interest
last_week = Array(9:15)
# Price predicted by the model
ŷ = f.(last_week)
# Compute rmse
rmse = sqrt(1/7 * (ŷ-Y)'*(ŷ-Y))
println("The RMSE in the last week is: $rmse USD")
```

Which is a pretty good error considering the order of magnitude of bitcoin price.

### Fourth grade Polynomial
In a similar way to the Parabola, the model proposed is 
$$y = g(x) = \beta_0 + \beta_1 + \beta_2 x^2 + \beta_3 x^3 +\beta_4 x^4$$
But now we are going to use the bigger dataframe with the bitcoin price over march, so the date ranges are 

| Purpose | Date Range | 
| --- | --- |
| Adjust model | 01-03 to 21-03 |
| Testing | 22-03 to 28-03 |

```julia
# Define arrays to adjust model
X = Array(1:21)
Y = Array(df_march[1:21,2])
# Define matrix to apply normal equation
M = [ones(21) X X.^2 X.^3 X.^4 ]
# Get the optimum parameters with normal equation
β = pinv(M)*Y

g(X)= β[1] + β[2]*X + β[3]*X^2 + β[4]*X^3 + β[5]*X^4 
month_days = Array(1:28)
 
plot(month_days, g.(month_days), color="green", label="Model", linewidth=2)
scatter!(X,Y, color="red", label="Adjusted Nodes")
plot!(month_days,df_march[:,2], color="darkblue", label="Real values", legend_position=(0.3,0.9))
```

It is evident this is not a good model for predicting the last week price, it could be due to overfitting giving by the many parameters that the model adjust comparing to the training data.

### Second order polinomial with 3 weeks trainning
In this case we will adjust a second order polynomial, but this time the model will train with the df_march.



```julia
X = Array(1:21)
Y = Array(df_march[1:21,2])
# Define matrix to apply normal equation
M = [ones(21) X X.^2 ]
# Get the optimum parameters with normal equation
β = pinv(M)*Y

g(X)= β[1] + β[2]*X + β[3]*X.^2 
month_days = Array(1:28)
 
plot(month_days, g.(month_days), color="green", label="Model", linewidth=2)
scatter!(X,Y, color="red", label="Adjusted Nodes")
plot!(month_days,df_march[:,2], color="darkblue", label="Real values", legend_position=(0.3,0.9))
```

```julia
y_real = Array(df_2[9:end,2])
y_pred = g.(Array(22:28))
rmse_3 = sqrt(1/7 * (y_pred-y_real)'*(y_pred-y_real))
println("rmse with third model: $rmse_3 USD")
```

Finally we conclude that the best model according to RMSE is the first, which is parabola with one week of training. simplicity is effective
